```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from describ import describing
```

```python
train_data = pd.read_csv("dataset_train.csv")
test_data = pd.read_csv("dataset_test.csv")
full_data = pd.concat([train_data, test_data])
full_data = full_data.drop(['Index'], axis=1)
full_data.head(4)
```

We have recomputed the results of the describe function in the previous script

We use the minimum and the maximum in the normalization of our data, so we create two functions.

```python
def minimum(frame):
    m = frame.iloc[0]
    for i in np.arange(1, len(frame)):
        if (m > frame.iloc[i]): m = frame.iloc[i]
    return m

def maximum(frame):
    m = frame.iloc[0]
    for i in np.arange(1, len(frame)):
        if (m < frame.iloc[i]): m = frame.iloc[i]
    return m

def normalize(frame):
    return (frame - minimum(frame))/(maximum(frame) - minimum(frame))
```

```python
end_results = {}

for name in full_data._get_numeric_data().columns:
    
    results = pd.DataFrame(index=['25%', '50%', '75%', '100%'])

    for house in full_data['Hogwarts House'].dropna().unique():
        
        col = pd.DataFrame(full_data.loc[full_data['Hogwarts House'] == house])
        col.columns = ['Feature']
        col['Feature'] = normalize(col['Feature'])
        col['Count'] = 1
        
        repartition = []
        for j in [0, 0.25, 0.5, 0.75]:
            rep = col.loc[(col['Feature'] >= j) & (col['Feature'] < (j + 0.25))]
            count = sum(rep['Count'])/sum(col['Count'])
            repartition.append(count)
            
        repartition = pd.DataFrame(repartition)
        repartition.index = ['25%', '50%', '75%', '100%']
        repartition.columns = [house]
        
        results = pd.concat([results, repartition], axis = 1)
        
    end_results[name] = results
```

```python
end_results
```

```python
for name in full_data._get_numeric_data().columns:
    np.transpose(end_results[name]).plot.bar(stacked=True, title=name)
```

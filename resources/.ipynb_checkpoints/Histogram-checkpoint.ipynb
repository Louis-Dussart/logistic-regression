```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_raw1 = pd.read_csv("dataset_train.csv")
data_raw2 = pd.read_csv("dataset_test.csv")
data_raw = pd.concat([data_raw1, data_raw2])
data_raw_1 = data_raw.drop(['Index'], axis=1)
data_raw_1.head()
```

We have recomputed the results of the describe function in the previous script

```python
description = data_raw_1.describe()
description
```

```python
columns = list(data_raw_1.columns)
columns = [e for e in columns if e not in ('Hogwarts House', 'First Name', 'Last Name', 'Birthday', 'Best Hand')]
```

```python
end_results = {}

for name in columns:
    
    results = pd.DataFrame(index=['25%', '50%', '75%', '100%'])

    for house in data_raw['Hogwarts House'].dropna().unique():
        new_data = data_raw_1.loc[data_raw['Hogwarts House'] == house]
        
        col = pd.DataFrame(new_data[name])
        col.columns = ['Feature']
        col['Feature'] = (col['Feature'] - description[str(name)].loc['min'])/(description[str(name)].loc['max'] - description[str(name)].loc['min'])
        
        col['Count'] = 1
        
        repartition = []
        for j in [0, 0.25, 0.5, 0.75]:
            rep = col.loc[(col['Feature'] >= j) & (col['Feature'] < (j + 0.25))]
            count = sum(rep['Count'])/sum(col['Count'])
            repartition.append(count)
            
        repartition = pd.DataFrame(repartition)
        repartition.index = ['25%', '50%', '75%', '100%']
        repartition.columns = [house]
        
        results = pd.concat([results, repartition], axis = 1)
        
    end_results[name] = results
```

```python
end_results
```

```python
for name in columns:
    np.transpose(end_results[name]).plot.bar(stacked=True, title=name)
```

We can see that in the Arithmancy and Care of Magical Creatures classes, houses have a similar grade repartition.

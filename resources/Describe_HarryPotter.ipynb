```python
import pandas as pd
import numpy as np
```

```python
data_raw1 = pd.read_csv("dataset_train.csv")
data_raw2 = pd.read_csv("dataset_test.csv")
data_raw = pd.concat([data_raw1, data_raw2])
data_raw.head()
```

We drop the index

```python
data_raw_1 = data_raw.drop(['Index'], axis=1)
data_raw_1.head()
```

We are going to manually code the following function

```python
data_raw_1.describe()
```

```python
columns = list(data_raw_1.columns)
columns = [e for e in columns if e not in ('Hogwarts House', 'First Name', 'Last Name', 'Birthday', 'Best Hand')]
columns
```

```python
results = pd.DataFrame(index=['Count','Mean','std','min','25%','50%','75%','max'])

for name in columns:
    col = pd.DataFrame(data_raw_1.filter([str(name)]))
    col.columns = ['Feature']
    col = col.dropna().reset_index(drop=True)
    col['One'] = 1
    
    count = sum(col['One'])
    mean = sum(col['Feature'])/count
    
    col['mean'] = mean
    col['std'] = (col['Feature'] - col['mean'])**2
    std = np.sqrt(1/count*sum(col['std']))
    
    minimum = col.Feature.iloc[0]
    for i in np.arange(1, len(col['Feature'])):
        if (minimum > col.Feature.iloc[i]): minimum = col.Feature.iloc[i]
            
    maximum = col.Feature.iloc[0]
    for i in np.arange(1, len(col['Feature'])):
        if (maximum < col.Feature.iloc[i]): maximum = col.Feature.iloc[i]
    
    percentage = pd.DataFrame(col.Feature)
    percentage.columns = ['Values']
    percentage = percentage.sort_values(by=['Values'], ascending = True, axis = 0).reset_index(drop=True)
    perc_25 = percentage.Values.iloc[int((len(percentage.Values))/4)]
    perc_50 = percentage.Values.iloc[int((len(percentage.Values))/2)]
    perc_75 = percentage.Values.iloc[int((len(percentage.Values))/4*3)]
    
    partial = pd.DataFrame([count, mean, std, minimum, perc_25, perc_50, perc_75, maximum])
    partial.index = ['Count','Mean','std','min','25%','50%','75%','max']
    partial.columns = [str(name)]
    results = pd.concat([results, partial], axis = 1)  
```
